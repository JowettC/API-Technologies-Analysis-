```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

csv_file_path = 'performance_script/results.csv'

df = pd.read_csv(csv_file_path)

# Extracting framework names
frameworks = df.columns.tolist()

# Metrics excluding "Throughout Analysis"
metrics = ["Get Request Duration", "Post Request Duration",
           "Put Request Duration", "Delete Request Duration",
           "Complex Query Duration (Less Data)", "Complex Query Duration (More Data)"]

# Plotting "Throughout Analysis"
plt.figure(figsize=(10, 6))
plt.bar(frameworks, df.loc[0], color=np.random.rand(3,))  # Assuming the first row is "Throughout Analysis"
plt.title('Throughout Analysis (Number of Requests)')
plt.xlabel('Framework')
plt.ylabel('Number of Requests')
plt.xticks(rotation=45)
plt.tight_layout()
plt.bar_label(plt.bar(frameworks, df.loc[0], color=np.random.rand(3,)))
plt.show()

# Create a figure for the all other metrics
plt.figure(figsize=(12, 8))
for i, metric in enumerate(metrics, start=1):  
    plt.plot(frameworks, df.iloc[i], label=metric, marker='o')
plt.title('API Performance Metrics (Excluding Throughout Analysis)')
plt.xlabel('Framework')
plt.ylabel('Seconds')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Creating individual graphs for each of the other metrics
for i, metric in enumerate(metrics, start=1):  
    plt.figure(figsize=(10, 6))
    plt.bar(frameworks, df.iloc[i], color=np.random.rand(3,))
    plt.title(metric)
    plt.xlabel('Framework')
    plt.ylabel('Seconds')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.bar_label(plt.bar(frameworks, df.iloc[i], color=np.random.rand(3,)))
    plt.show()


```

```python
# filter only frameworsk with FASTAPI in the name
df_fastapi = df.filter(regex='FastAPI')
df_fastapi = df_fastapi.drop([0])
headers = df_fastapi.columns
# combine all rows of data and take average
df_fastapi = df_fastapi.mean()
# plot the results
plt.figure(figsize=(10, 6))
plt.bar(headers, df_fastapi, color=np.random.rand(3,))
plt.title('FastAPI Performance')
plt.xlabel('Metric')
plt.ylabel('Seconds')
plt.xticks(rotation=45)
plt.tight_layout()
plt.bar_label(plt.bar(headers, df_fastapi, color=np.random.rand(3,)))
plt.show()
```
